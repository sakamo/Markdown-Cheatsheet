## 練習問題6.1

以下の内容でじゃんけんプログラムを作成してください

randomモジュールをインポート
0から9の整数をランダムに生成し、変数jankenに入れる

```python
import random

janken = random.randint(0,5)
if janken == 0:
    print("グーで勝ち")
elif janken == 1:
    print("チョキで勝ち")
elif janken == 2:
    print("パーで勝ち")
elif janken == 3:
    print("グーで負け")
elif janken == 4:
    print("チョキで負け")
elif janken == 5:
    print("パーで負け")
else:
    print("あいこ")
```

## 練習問題6.2

練習6.1で作成したじゃんけんプログラムの条件を下記に従って変更してください

条件
jankenが5より大きい　"勝ち"と出力
jankenが5　"あいこ"と出力
jankenが4以下　"負け"と出力

```python
import random

janken = random.randint(0,9)
if janken > 5:
    print("勝ち")
elif janken == 5:
    print("あいこ")
else :
    print("負け")
```
